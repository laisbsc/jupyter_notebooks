#### 2. Plot with number of beer, spirit and wine servings for each of the top 5 countries   
To illustrate the details of the amount and type of alcohol served on each country, a grouped horizontal bar plot was used, which shows the countries against the consumption of each type of alcohol.

```python
# new dataframe without the 'total_alcohol_consumption' column
df_servings = df_rank.filter(items=['country', 'beer_servings', 'spirit_servings', 'wine_servings'])
```

```python
# plot the number or beer, spirit and wine servings for each of the top 5 countries
ax = df_servings.plot(x='country', kind='barh', lw=2, colormap='rainbow',
                subplots=False, layout=(2,2),
                figsize=(8,6), title='Bar plot decripting the servings\n of drinks for the top countries on alcohol consumption')
# adding names to axis
ax.set(xlabel='Alcohol Servings', ylabel='Country')
```

From the plot above it is possible to conclude that among the countries with the highest alcohol consumption in the world, **Grenada** and **Belarus** are by far the ones who serve the most spirits. The countries where beers are most common are **Czech Republic** and **Lithuania**. And finally, **Andorra** is the only territory where wine is more consumed than any other alcoholic drink.
On the other hand, **Grenada** is the country to consume the least servings on wine, followed by **Belarus** and **Lithuania**.

#### 3. Plot with distribution of the total liters of alcohol consumption per (top 5) country
The distribution indicates the number of similar occurrences in each column of a certain attribute, i.e. the frequency of certain values on a dataset. In this case, it is requested to generate a plot based on the top 5 countries consumers of alcohol in the world. The dataset is small but a histogram can help analyse the distribution of the data.

```python
# create a dataframe with country and total alcohol columns only
df_total = df_rank.filter(items=['country', 'total_litres_of_pure_alcohol'])
df_total.head()
```

```python
df_total.plot(x='country', y='total_litres_of_pure_alcohol', kind='hist', 
                subplots=False, layout=(2,2),
                figsize=(8,6), title='Histogram illustrating the total\n alcohol consumption (in liters) per country')
```

Based on the plot above, it is possible to observe that most countries (among the 5 top) are around the 12 liters alcohol consumption mark, the peak. The remaining (3 countries) are placed on the 12.5, 13 and 14 liters mark per year. No countries are on the marks 13.5 and 14 liters.
It is relevant to observe that distribution plots (histograms, KDEs, ECDFs, etc.) are more representative the bigger the dataset they represent. To illustrate such difference, the plot graph is a histogram plot with the raw (unfiltered) dataset (`df_raw`).

```python
# [non-required] plot the distribution of alcohol per country in the entire dataset (all countries)
df_raw.plot(x='country', y='total_litres_of_pure_alcohol', kind='hist',
                subplots=False, layout=(2,2),
                figsize=(8,6), title='Histogram illustrating the total\n alcohol consumption (in liters) per country')
```

As imagined, from the plot above it is possible to see that most countries in the world are between the zero and below two liters of alcohol consumed per year. The second most common consumption rate is between 3 and 6 liters and 7.5 to just above 10 liters.

#### 4. Continent which consumes the largest `total liters of pure alcohol`
To illustrate the continents which consume the most alcohol, the student decided to used a pie plot. PIe plots are used to demonstrate the proportions of categories on a dataset. Since the number os continents is limited to six and there are significant differences between each portion, a pie chart is suitable for this use-case.

```python
# new dataframe with only 'continent' and 'total_alcohol_consumption' columns
df_cont = df_raw.filter(items=['continent', 'total_litres_of_pure_alcohol'])

# sum the alcohol consumption on the grouped rows (selected by continents)
df_continent = df_cont.groupby('continent', as_index=True).agg({"total_litres_of_pure_alcohol": "sum"})
df_continent
```

```python
# plot a pie chart to show the proportions of alcohol consumption by each continent
df_continent.plot(y='total_litres_of_pure_alcohol', kind='pie', colormap='tab20c',
                subplots=False, layout=(2,2), ylabel='', autopct='%.2f',
                figsize=(10,8), title='Pie chart illustrating the percentage (%) \n'+
                                          'proportion of alcohol consumption per continent')
```

It is possible to infer from the graph above that Europe is by far the biggest consumer of alcohol among all continents with almost half of the consuming market. Followed by Africa and North America on a very similar proportion. In contrast, Oceania has the lowest consumption of alcohol (54 liters per year), followed by South America (76 liters per year or around 8% of the market) and Asia with almost double consumption if compared to the Pacific Ocean continent.  
Further analysis to find the cause for the Old Continent's high alcohol consumption rates should take into consideration the weather in the top consumer regions (countries) and cultural differences. Moreover, it is possible to observe from the illustration below that Europe has the highest consumption of wine in the entire world, which contributes to the overall result of the plot above.

#### 5. Average `beer` and `wine` servings by continent
To illustrate the comparison these two discrete categories, a vertical grouped bar plot was used. The aim was to illustrate the rates of beer and wine consumption on the different continents. 

```python
# new dataframe with only 'continent' and 'beer_servings' and 'wine_servings' columns
df_serv = df_raw.filter(items=['continent', 'beer_servings', 'wine_servings'])

# sum the alcohol consumption on the grouped rows (selected by continents)
df_beer_wine = df_serv.groupby('continent', as_index=True).agg({"beer_servings": "mean", "wine_servings":"mean"})
df_beer_wine
```

```python
# [non-required] plot with subplots > not grouped bars but one for beer and another for wine
# uncomment and run to generate separate subplots
# df_servings.plot(kind='bar')
# df_beer_wine.plot(kind='bar', colormap='tab20c',
#                 subplots=True, layout=(2,2), xlabel='Continent', ylabel='Servings',
#                 figsize=(10,8), title='Bar plots illustrating the average (mean)\n beer and wine servings per continent')
```

```python
# grouped bar plot
df_beer_wine.plot(kind='bar', colormap='rainbow',
                subplots=False, layout=(2,2), xlabel='Continent', ylabel='Servings',
                figsize=(10,8), title='Bar plot illustrating the average (mean) beer and wine servings per continent')
```

The plot above illustrates the beer and wine number of servings on each continent in the world. It is possible to observe that Europe has the highest alcohol consumption overall but also the most significant wine consumption numbers than anywhere else in the world. South America consumes a significant volume of beer but has a much less significant  'wine culture'. All continents consume more beer servings than wine and Asia is the continent with the least significant servings of both beer and wine in the world. Although, it would be interesting to check if 'sake' was logged into the dataset under the 'wine' category, which could explain the disparity.
